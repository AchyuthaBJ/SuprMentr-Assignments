## sum until user stops: keep asking the user to enter numbers and calculate the sum. stop only when the user enters 0

```python
Sum = 0
while True:
    enter_num = int(input("Enter your number. Enter 0 to exit"))
    if enter_num == 0:
        break
    Sum += enter_num

print("The sum is",Sum) 
```
