# Numpy array of temperatures 
print max min avg and last 3

```python
import numpy as np
array = [28,32,30,37,36,38]
temp = np.array(array)

print("max and min temperatures are ",max(temp),min(temp))
print("Average temperature is ",np.average(temp))
print("last 3 days temp are: ", temp[len(temp)-3:])
```
