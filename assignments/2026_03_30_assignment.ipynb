# Resume Prompt
--Weak prompt: <br>
Write a resume for me <br>
--Strong prompt : <br>
Create a professional resume for a Computer Science student skilled in Python, Machine Learning, and Web Development. Include sections: Summary, Skills, Projects, Education, and Certifications. Keep it concise and ATS-friendly.

# Business Idea Prompt
--Weak Prompt: <br>
Give me a business idea<br>
--Strong Prompt : <br>
Suggest 3 innovative micro-SaaS business ideas suitable for a college student with basic programming skills. Each idea should include target audience, problem solved, and monetization strategy.

# Study Plan Prompt
--Weak Prompt: <br>
Make a study plan <br>
--Strong Prompt:<br>
Create a 4-week study plan to learn Data Structures and Algorithms for placement preparation. Allocate 2 hours daily and include topics, practice questions, and revision days.
