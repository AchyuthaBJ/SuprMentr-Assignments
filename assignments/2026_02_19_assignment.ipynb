# Student Data Manager

```python
import pandas as pd

data = pd.DataFrame({
    "Name": ["Aarav", "Riya", "Karan", "Sneha", "Arjun"],
    "Maths": [85, 92, 76, 88, 90],
    "Science": [78, 89, 74, 91, 84],
    "English": [82, 94, 70, 87, 88],
    "Social": [80, 86, 72, 85, 83],
    "Computer": [90, 95, 79, 93, 89]
})
data["Total"] = data["Maths"]+data["Science"]+data["English"]+data["Social"]+data["Computer"]
data["Average"] = data["Total"]/5

grade = []
for x in data["Average"]:
    if x>=90:
        grade.append("A")
    elif 80<=x<90:
        grade.append("B")
    elif 70<=x<80:
        grade.append("C")

grade
data["Grade"] = grade

                            
```
