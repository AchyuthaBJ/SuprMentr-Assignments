# Fizz-Buzz

```python
def fizzbuzz(n):
    fb=0
    f=0
    b=0
    for i in range(1,n+1):
        if i%3==0 and i%5==0:
            print("FizzBuzz")
            fb+=1
        elif i%3==0:
            print("Fizz")
            f+=1
        elif i%5==0:
            print("Buzz")
            b+=1
        else:
            print(i)
    
    return fb,f,b

Fizzbuzz,fizz,buzz = fizzbuzz(20)
print(Fizzbuzz,fizz,buzz)
```
