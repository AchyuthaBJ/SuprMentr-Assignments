```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "alice", "Charlie", None],
    "Age": [25, np.nan, 25, 30, 30],
    "City": ["New York", "Los Angeles", "new york", "Chicago", "Chicago"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Fill missing Age with mean
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Drop rows where Name is missing
df = df.dropna(subset=["Name"])
```

```python
df = df.drop_duplicates()
df = df.drop_duplicates(subset=["Name", "City"])
# Convert text to lowercase
df["Name"] = df["Name"].str.lower()
df["City"] = df["City"].str.lower()

# Remove extra spaces
df["Name"] = df["Name"].str.strip()
df["City"] = df["City"].str.strip()
print(df)
```

1. Improves Model Accuracy:
    Dirty data (missing, inconsistent values) leads to wrong patterns
    Models trained on clean data perform better and generalize well
2. Ensures Consistency:
    "New York" vs "new york" → treated as different values
    Standardization avoids duplicate categories
3. Prevents Bias & Errors:
    Missing values can skew results
    Duplicates can overweight certain data points
4. Better Data Interpretation:
    Clean data → easier visualization & analysis
    Makes insights trustworthy
5. Required for ML Pipelines:
    Most algorithms (in scikit-learn) cannot handle missing values directly
