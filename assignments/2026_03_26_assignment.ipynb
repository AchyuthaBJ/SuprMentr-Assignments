```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download VADER lexicon (run once)
nltk.download('vader_lexicon')

# Initialize analyzer
sia = SentimentIntensityAnalyzer()

# Function to classify sentiment
def analyze_sentiment(review):
    score = sia.polarity_scores(review)

    if score['compound'] >= 0.05:
        return "Positive 😊"
    elif score['compound'] <= -0.05:
        return "Negative 😞"
    else:
        return "Neutral 😐"

# Test on 5 movie reviews
reviews = [
    "This movie was absolutely wonderful and inspiring",
    "I did not like this movie at all, it was boring",
    "The storyline was okay, nothing special",
    "Terrible acting and very disappointing film",
    "Amazing visuals and brilliant performance"
]

# Output results
print("🎬 Sentiment Analysis Results:\n")

for i, review in enumerate(reviews, 1):
    sentiment = analyze_sentiment(review)
    print(f"{i}. Review: {review}")
    print(f"   Sentiment: {sentiment}\n")
```
