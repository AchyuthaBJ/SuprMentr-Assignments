# Assignment Name : Semantic Meaning
# Description : Find 5 word pairs and explain semantic similarity.

1. Happy – Joyful <br>
--Similarity Type: Synonyms<br>
--Explanation: Both words express a state of pleasure or happiness. “Joyful” is slightly stronger emotionally, but the meanings largely overlap.<br>
2. Car – Vehicle<br>
--Similarity Type: Hierarchical (Hypernym)<br>
--Explanation: A car is a type of vehicle. So, “vehicle” is a broader category that includes cars, buses, trucks, etc.<br>
3. Teacher – Educator<br>
--Similarity Type: Synonyms / Contextual<br>
--Explanation: Both refer to someone who teaches or imparts knowledge, though “educator” is more formal and broader.<br>
4. Hot – Cold<br>
--Similarity Type: Antonyms<br>
--Explanation: These words have opposite meanings, so their semantic similarity is very low (or negative).<br>
5. Dog – Puppy<br>
--Similarity Type: Related (Age-based relation)<br>
--Explanation: A puppy is a young dog, so they are closely related but not identical in meaning.
