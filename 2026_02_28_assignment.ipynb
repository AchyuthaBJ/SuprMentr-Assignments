```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("data.csv")
data.tail()
data["Driven_Wheels"].value_counts()
```

```python
x = data["Driven_Wheels"].value_counts()
types = x.index.to_list()
values = x.values.tolist()
plt.bar(types,values)
plt.title("Drive Type count")
plt.xlabel("Drive Type")
plt.ylabel("Count")
plt.show()
```

```python
import numpy
x = (data["Transmission Type"].value_counts())
types = x.index.to_list()
values = x.values.tolist()

plt.pie(values,labels=types)
plt.legend(loc=1)
plt.show()
```
